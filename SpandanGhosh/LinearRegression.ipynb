```python
import numpy as np
import pandas as pd

df_train=pd.read_csv('C:/Users/712742/Desktop/train.csv')#read csv train file
df_test=pd.read_csv('C:/Users/712742/Desktop/test.csv')#read csv test file

x_train=df_train['x']#take x training values as a list
y_train=df_train['y']#take y training values as a list
x_test=df_test['x']#take x test values as a list
y_test=df_test['y']#take y test values as a list

x_train=np.array(x_train)#take x training values as a numpy array
y_train=np.array(y_train)#take y training values as a numpy array
x_test=np.array(x_test)#take x test values as a numpy array
y_test=np.array(y_test)#take x test values as a numpy array

#x_train and x_test are arrays(one dimentional) containing n elements

x_train=x_train.reshape(-1,1)#converts into nx1 array(two dimensional)
x_test=y_test.reshape(-1,1)#converts into nx1 array(two dimensional)

n=700        #number of data items
alpha=0.0000001#learing rate
epoch=0      #number of iteration
a_0=0        #initial value of c where y=mx+c
a_1=0        #initial value of m where y=mx+c

while(epoch<3000):                      
    #when epoch is 1000, the values of a_0,a_1 are so small that nan is printed, works fine when epoch is 100 it works fine
    #To solve above problem, ihave altered the values of alpha and epoch
    y_pred=a_0+a_1*x_train
    error=y_pred-y_train
    a_0=a_0-alpha*2*np.sum(error)/n
    a_1=a_1-alpha*2*np.sum(error*x_train)/n
    epoch+=1
    
print('Mean Square Error is '+str(np.sum(error**2)/n)) # mean square error 
print('The straight line equation is y=mx+c')
print('The value of c is '+str(a_0))
print('The value of m is '+str(a_1))

y_prediction = a_0 + a_1 * x_test
print('R2 Score:',r2_score(y_test,y_prediction))
```
